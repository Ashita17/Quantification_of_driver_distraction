```python
#import dependencies

import cv2
from matplotlib import pyplot as plt
import numpy as np
import imutils
import easyocr
import pytesseract
from skimage import io, img_as_float
from skimage.metrics import peak_signal_noise_ratio
from scipy import ndimage as nd
from skimage.filters import median
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing import image
from IPython.display import Image
import matplotlib.pyplot as plt
```

```python
IMAGE_PATH = '/Users/ashitasrivastava/Desktop/driving_images/ESRGAN modified/image14_rlt.png'
```

```python
im=cv2.imread(IMAGE_PATH)

#getting the width and height of the image
# height,width=im.shape[:2]
# print(f"Width: {width}, Height: {height}")

roi1=cv2.selectROI('selectroi1',im)
roi2=cv2.selectROI('selectroi2',im)

im_cropped1 =im[int(roi1[1]):int(roi1[1]+roi1[3]), #starting fropm the y coordinate of the top left corner and sweeping across all the rows through the height
			int(roi1[0]):int(roi1[0]+roi1[2])]
im_cropped2 =im[int(roi2[1]):int(roi2[1]+roi2[3]), #starting fropm the y coordinate of the top left corner and sweeping across all the rows through the height
			int(roi2[0]):int(roi2[0]+roi2[2])]

cv2.destroyAllWindows()
cv2.waitKey(2000)
```

```python
#Reading in grayscale 

# img = cv2.imread('/Users/ashitasrivastava/Desktop/driving_images/ESRGAN modified/image14_rlt.png')
gray1 = cv2.cvtColor(im_cropped1, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray1, cv2.COLOR_BGR2RGB))
```

```python
reader = easyocr.Reader(['en'])
result1 = reader.readtext(gray1)
result1[0][1]
```

```python
extracted_text1 = pytesseract.image_to_string(im_cropped1)
extracted_text1
```

```python
gray2 = cv2.cvtColor(im_cropped2, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray2, cv2.COLOR_BGR2RGB))
```

```python
result2 = reader.readtext(gray2)
result2[0][1]
```

```python
extracted_text2 = pytesseract.image_to_string(gray2)
extracted_text2
```
