```python
import os
import easyocr
import cv2
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from pathlib import Path
from PIL import Image, ImageEnhance
```

### Creating Dataframe

```python
# Generate random data
Speed_Data = {
    'Time Frame': np.random.rand(4),  # Random values for column A
    'Speed': np.random.rand(4)   # Random values for column B
}

# Create DataFrame
df = pd.DataFrame(Speed_Data)

print(df)
```

### Extracting data from the images

```python
#path

path='/content/drive/MyDrive/projects/driver distraction and correlation with speed/images'
images = os.listdir(path)
```

```python
#confirming if the loop is working well

folder_path ='/content/drive/MyDrive/projects/driver distraction and correlation with speed/images'
images = os.listdir(folder_path)

for filename in images:
    filepath = os.path.join(folder_path, filename)

    print(filename)
```

```python
i=0
for image in images:

    if image != '.DS_Store':

    #reading images in the folder with ERSGAN modified images
        im=cv2.imread(os.path.join(path,image))

        #selecting the roi for text extraction

        roi1=cv2.selectROI('selectroi1',im)
        roi2=cv2.selectROI('selectroi2',im)

        im_cropped1 =im[int(roi1[1]):int(roi1[1]+roi1[3]), #starting fropm the y coordinate of the top left corner and sweeping across all the rows through the height
                int(roi1[0]):int(roi1[0]+roi1[2])]
        im_cropped2 =im[int(roi2[1]):int(roi2[1]+roi2[3]), #starting fropm the y coordinate of the top left corner and sweeping across all the rows through the height
                int(roi2[0]):int(roi2[0]+roi2[2])]

        cv2.destroyAllWindows()
        cv2.waitKey(2000)

        #Image processing: converting to grayscale and increasing contrast

        gray1 = cv2.cvtColor(im_cropped1, cv2.COLOR_BGR2GRAY)

        image_pil1 = Image.fromarray(cv2.cvtColor(gray1, cv2.COLOR_BGR2RGB))
        enhancer = ImageEnhance.Contrast(image_pil1)
        enhanced_image_pil1 = enhancer.enhance(2.5)
        enhanced_image1 = cv2.cvtColor(np.array(enhanced_image_pil1), cv2.COLOR_RGB2BGR)

        # plt.imshow(cv2.cvtColor(gray1, cv2.COLOR_BGR2RGB))

        reader = easyocr.Reader(['en'])

        result1 = reader.readtext(enhanced_image1)
        Result1=result1[0][1]
        print(Result1)

        gray2 = cv2.cvtColor(im_cropped2, cv2.COLOR_BGR2GRAY)

        image_pil2 = Image.fromarray(cv2.cvtColor(gray2, cv2.COLOR_BGR2RGB))
        enhancer = ImageEnhance.Contrast(image_pil2)
        enhanced_image_pil2 = enhancer.enhance(2.5)
        enhanced_image2 = cv2.cvtColor(np.array(enhanced_image_pil2), cv2.COLOR_RGB2BGR)

        # plt.imshow(cv2.cvtColor(gray2, cv2.COLOR_BGR2RGB))

        result2 = reader.readtext(gray2)
        Result2=result2[0][1]
        print(Result2)

        #Adding the results to the dataframe

        df.loc[i]=[Result1,Result2]
        i+=1
```
