```python
import os
import easyocr
import cv2
import pytesseract
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from skimage import io, img_as_float
from skimage.metrics import peak_signal_noise_ratio
from scipy import ndimage as nd
from skimage.filters import median
from pathlib import Path
```

```python
im=cv2.imread('/content/WhatsApp Image 2024-01-12 at 11.48.46_76ff6c62.jpg')
# roi1=cv2.selectROI('selectroi1',im)
# im_cropped1 =im[int(roi1[1]):int(roi1[1]+roi1[3]), #starting fropm the y coordinate of the top left corner and sweeping across all the rows through the height
# 			int(roi1[0]):int(roi1[0]+roi1[2])]
# cv2.destroyAllWindows()
# cv2.waitKey(2000)
gray1 = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray1, cv2.COLOR_BGR2RGB))
reader = easyocr.Reader(['en'])

result1 = reader.readtext(gray1)
Result1=result1[0][1]
print(Result1)
```

```python
from PIL import Image, ImageEnhance
import easyocr

image_pil = Image.fromarray(cv2.cvtColor(gray1, cv2.COLOR_BGR2RGB))

# Convert Pillow Image to ImageEnhance object for contrast adjustment
enhancer = ImageEnhance.Contrast(image_pil)

# Adjust contrast - increase by a factor (2.0 for example)
enhanced_image_pil = enhancer.enhance(2.5)  # Adjust the factor as needed

# Convert back to numpy ndarray (OpenCV format)
enhanced_image = cv2.cvtColor(np.array(enhanced_image_pil), cv2.COLOR_RGB2BGR)

# Display or save the enhanced image using OpenCV
plt.imshow(enhanced_image)
cv2.destroyAllWindows()
cv2.waitKey(2000)
```
